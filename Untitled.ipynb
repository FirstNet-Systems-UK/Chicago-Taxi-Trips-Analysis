## preprocessing

```python
# Convert date and time columns to datetime format
date_time_columns = ['Trip Start Month', 'Trip End Month', 'Trip Start Time', 'Trip End Time']
df[date_time_columns] = df[date_time_columns].apply(pd.to_datetime)

# Handle missing values (example: fill missing numerical values with mean)
numerical_columns = ['Trip Minutes', 'Trip Miles', 'Trip Kilometers', 'Fare', 'Tips', 'Tolls', 'Extras', 'Trip Total']
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].mean())

# Encode categorical variables (example: one-hot encoding for Payment Type)
df = pd.get_dummies(df, columns=['Payment Type'])

# Feature engineering (example: extract day of the week from Trip Start Time)
df['Day of Week'] = df['Trip Start Time'].dt.day_name()

# Normalize numerical features (example: Min-Max scaling)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

# Drop unnecessary columns if needed
df = df.drop(['Taxi ID', 'Pickup Census Tract', 'Dropoff Census Tract', 'Pickup Location', 'Dropoff Location'], axis=1)

# Check the updated DataFrame
print(df.head())
```

## Descriptive Statistics and Visual Techniques


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Summary statistics
summary_stats = df.describe()
print(summary_stats)
```

```python

# Visualizations
# Histogram of Fare distribution
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Fare', bins=30, kde=True)
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
plt.savefig(" Histogram of Fare distribution")

```

```python
# Selecting relevant columns for the box plot
payment_columns = ['Payment Type_Cash', 'Payment Type_Credit Card', 'Payment Type_Mobile', 'Payment Type_Unknown']

# Melt the DataFrame to plot all payment types in a single box plot
melted_df = df.melt(value_vars=payment_columns, var_name='Payment Type', value_name='Payment Value')

# Box plot of Trip Minutes by Payment Type
plt.figure(figsize=(12, 8))
sns.boxplot(data=melted_df, x='Payment Type', y='Payment Value')
plt.title('Trip Minutes by Payment Type')
plt.xlabel('Payment Type')
plt.ylabel('Trip Minutes')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent overlapping labels
plt.savefig("Box plot of Trip Minutes by Payment Type")  # Save the plot
plt.show()
```

```python
# Scatter plot of Trip Miles vs. Fare
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Trip Miles', y='Fare')
plt.title('Trip Miles vs. Fare')
plt.xlabel('Trip Miles')
plt.ylabel('Fare')
plt.show()
plt.savefig("Scatter plot of Trip Miles vs Fare")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your DataFrame containing the relevant data

# Scatter plot of Trip Miles vs. Fare
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Trip Miles', y='Fare')
plt.title('Trip Miles vs. Fare')
plt.xlabel('Trip Miles')
plt.ylabel('Fare')
plt.savefig("Scatter plot of Trip Miles vs Fare.png")  # Save the plot
plt.show()

# Descriptive Insights
# Calculate correlation between Trip Miles and Fare
correlation = df['Trip Miles'].corr(df['Fare'])

# Identify outliers
outliers = df[(df['Fare'] > df['Fare'].quantile(0.95)) | (df['Fare'] < df['Fare'].quantile(0.05))]

# Data Distribution Insights
# You can analyze clusters or patterns in the scatter plot based on trip miles and fare

# Print descriptive insights
print("Descriptive Insights:")
print(f"Correlation between Trip Miles and Fare: {correlation:.2f}")
print("Number of Outliers:", len(outliers))
```
